```python
import pandas as pd
```

```python
excel_path = r"C:\Users\Victoria\Data Engineering\Pandas\world_population_excel_workbook.xlsx"
df = pd.read_excel(excel_path, sheet_name="Sheet1")
df
```

```python
# fiilter by value

df[df['Rank'] < 10]
```

```python
#fiilter to specific values

specific_countries = ['Bangladesh', 'Brazil']

df[df['Country'].isin(specific_countries)]
```

```python
# partial search, similar to like in SQL

df[df['Country'].str.contains('United')]
```

```python
# set an index  based on the country
df2 = df.set_index('Country')

df2
```

```python
# show only the columns that we want to see, axis will filter on the x or y

df2.filter(items = ['Capital', 'CCA3'], axis = 1)
```

```python
#filter to an x axis index
df2.filter(items = ['Zimbabwe'], axis = 0)
```

```python
#partial filter to an x axis index
df2.filter(like = 'United', axis=0)
```

```python
# get a row in a data frame by its index value
df2.loc['United States']
```

```python
# get a row in a data frame by its index value
df2.iloc[3]
```

```python
#sort by a column
df[df['Rank']<10].sort_values(by='Rank', ascending=False)
```

```python
#sort by multiple columns
df[df['Rank']<10].sort_values(by=['Capital', 'Country'], ascending=[False, True])
```
