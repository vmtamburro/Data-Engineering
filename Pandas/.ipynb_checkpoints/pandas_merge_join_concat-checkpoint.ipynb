```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Victoria\Data Engineering\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Victoria\Data Engineering\Pandas\LOTR 2.csv")
df2
```

```python
#inner join where the keys are the same

df1.merge(df2, how="inner", on=["FellowshipID", "FirstName"])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
#takes each value in the left data fame, and crosss each value in the right data frame
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
#join function works better if we have an index. Syntax-wise, merge is easier
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

```python
#similar to a union
pd.concat([df1,df2], join = 'outer', axis = 1)
```
